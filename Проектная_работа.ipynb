```python
# Рассчитаем цену за м² - Это будет - Monetary (деньги)
df['price_per_m2'] = df['price'] / df['total_meters']

# Разделим на 3 группы (1 — дешевые, 3 — дорогие)
df['M_score'] = pd.qcut(df['price_per_m2'], q=3, labels=[1, 2, 3])
```

```python
# Меняем последнюю и предпоследнюю колонку местами, чтобы значения R, F и M-score были рядом в таблице
price_per_m2, M_score = "price_per_m2", "M_score"

# Сохраняем данные из колонок
data_price_per_m2 = df[price_per_m2].copy()
data_M_score = df[M_score].copy()

# Меняем данные местами
df[price_per_m2] = data_M_score
df[M_score] = data_price_per_m2

# Меняем названия колонок
df = df.rename(columns={price_per_m2: M_score, M_score: price_per_m2})
```

```python
# меняем тип данных category на числовой
df['R_score'] = pd.to_numeric(df['R_score'], errors='coerce')
df['F_score'] = pd.to_numeric(df['F_score'], errors='coerce')
df['M_score'] = pd.to_numeric(df['M_score'], errors='coerce')
```

R (Recency) = средняя "свежесть" объявлений в районе,

F (Frequency) = количество объявлений (активность рынка для каждого района),

M (Monetary) = средняя цена за м² (цена за метр квадратный в районе и платёжеспособность людей для данного района).

```python
#1. Агрегация данных по районам
rfm_districts = df.groupby('district').agg(
    R_score=('R_score', 'mean'),     # Средний R_score (свежесть)
    F_score=('row_number', 'count'), # Количество объявлений (Frequency)
    M_score=('price_per_m2', 'mean') # Средняя цена за м² (Monetary)
).reset_index()

# Переименуем для ясности
rfm_districts = rfm_districts.rename(columns={'F_score': 'ads_count'})
```

```python
#2. Нормализация оценок (шкала 1-3)
# Recency: чем выше R_score, тем лучше (делим на 3 квартиля)
rfm_districts['R_norm'] = pd.qcut(
    rfm_districts['R_score'], 
    q=3, 
    labels=[1, 2, 3]
)

# Frequency: чем больше объявлений, тем лучше
rfm_districts['F_norm'] = pd.qcut(
    rfm_districts['ads_count'], 
    q=3, 
    labels=[1, 2, 3]
)

# Monetary: чем выше цена, тем лучше
rfm_districts['M_norm'] = pd.qcut(
    rfm_districts['M_score'], 
    q=3, 
    labels=[1, 2, 3]
)
```

```python
#3. Создание RFM-сегментов
# Комбинированный сегмент (например, '3-3-3')
rfm_districts['RFM_segment'] = (
    rfm_districts['R_norm'].astype(str) + '-' + 
    rfm_districts['F_norm'].astype(str) + '-' + 
    rfm_districts['M_norm'].astype(str)
)

# Общий RFM-балл (сумма)
rfm_districts['RFM_score'] = (
    rfm_districts['R_norm'] + 
    rfm_districts['F_norm'] + 
    rfm_districts['M_norm']
)

# Категоризация районов
def get_rfm_group(score):
    if score >= 8:
        return 'Премиум-районы'
    elif score >= 6:
        return 'Перспективные'
    elif score >= 4:
        return 'Средние'
    else:
        return 'Низкий приоритет'

rfm_districts['RFM_group'] = rfm_districts['RFM_score'].apply(get_rfm_group)
```

```python
import geopandas as gpd
import folium
import numpy as np
from folium.plugins import MiniMap, Fullscreen
from IPython.display import display

# 1. Загрузка данных
geo_data = gpd.read_file(r"C:\Users\yaaegorov\Documents\OTUS\Moscow.geojson")

# 2. Подготовка данных
if 'Weighted_RFM' not in rfm_districts.columns:
    rfm_districts['Weighted_RFM'] = (rfm_districts['R_norm']*0.3 + 
                                    rfm_districts['F_norm']*0.2 + 
                                    rfm_districts['M_norm']*0.5)

# 3. Слияние данных и обработка цен
merged_data = geo_data.merge(rfm_districts, on='district', how='left')

# Преобразование цен в целые числа
merged_data['M_score'] = (
    merged_data['M_score']
    .replace([np.inf, -np.inf], np.nan)  # Обрабатываем бесконечности
    .fillna(0)  # Заполняем пропуски нулями
    .round()    # Округляем
    .astype('int64')  # Преобразуем в целые числа
)

# 4. Создание карты с улучшенными параметрами
moscow_map = folium.Map(
    location=[55.7558, 37.6173], 
    zoom_start=11,
    tiles='cartodbpositron',
    width='100%', 
    height='600px'
)

# 5. Добавляем Choropleth с цветовой заливкой
choropleth = folium.Choropleth(
    geo_data=merged_data,
    data=merged_data,
    columns=['district', 'Weighted_RFM'],
    key_on='feature.properties.district',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.3,
    legend_name='RFM Индекс (чем выше - тем лучше)',
    highlight=True,
    bins=5,
    reset=True
).add_to(moscow_map)

# 6. Добавляем подсказки с явным форматированием чисел
style_function = lambda x: {
    'fillColor': '#ffffff',
    'color': 'black',
    'weight': 1,
    'fillOpacity': 0.1
}

folium.GeoJson(
    merged_data,
    style_function=style_function,
    tooltip=folium.features.GeoJsonTooltip(
        fields=['district', 'M_score', 'ads_count', 'Weighted_RFM'],
        aliases=['Район:', 'Цена за м² (руб):', 'Кол-во объявлений:', 'RFM Индекс:'],
        formatters={
            'M_score': '{:,.0f}',      # Цены как целые числа
            'Weighted_RFM': '{:.2f}',  # RFM с 2 знаками после запятой
            'ads_count': '{:,.0f}'     # Количество объявлений как целое
        },
        style=("background-color: white; color: #333333; font-family: arial; font-size: 12px; padding: 5px;"),
        localize=True
    )
).add_to(moscow_map)

# 7. Добавляем плагины
MiniMap(toggle_display=True).add_to(moscow_map)      #Добавляет миниатюрную карту-помощник (мини-карту) в угол основной карты.
Fullscreen(position='topright').add_to(moscow_map)    #Параметр toggle_display=True позволяет скрывать/показывать мини-карту кликом по иконке.

# 8. Отображаем карту в Jupyter
display(moscow_map)

```
